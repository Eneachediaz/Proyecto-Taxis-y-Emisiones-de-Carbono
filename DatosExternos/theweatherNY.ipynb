**Diccionario:**
* STATION: estación de la medición realizada.
* NAME: nombre de la estación de medición.
* DATE: fecha de la medición realizada.
* AWND: Velocidad promedio del viento (en millas por hora).
* PRCP: Precipitación acumulada (en pulgadas).
* SNOW: Cantidad de nieve caída (en pulgadas).
* SNWD: Profundidad de la nieve en el suelo (en pulgadas).
* TAVG: Temperatura promedio diaria (promedio de máxima y mínima) en grados Fahrenheit.
* TMAX: Temperatura máxima diaria en grados Fahrenheit.
* TMIN: Temperatura mínima diaria en grados Fahrenheit.
* WT01, WT02, WT03, etc.: Códigos de tiempo meteorológico (por ejemplo, niebla, lluvia, tormenta, etc.). Estos códigos indican condiciones climáticas específicas.
