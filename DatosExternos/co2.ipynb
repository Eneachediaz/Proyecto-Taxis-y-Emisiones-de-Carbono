```python
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' con las columnas 'year' y 'total_co2_emissions'
# Puedes ajustar los nombres de las columnas según tu DataFrame real

# Crear una figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Configurar los datos para las barras
x = df_nuevo['year']
y = df_nuevo['total_co2_emissions']

# Graficar las barras
ax.bar(x, y, color='b', alpha=0.7)

# Configurar etiquetas y título
ax.set_xlabel('Año')
ax.set_ylabel('Emisiones de CO2 (toneladas)')
ax.set_title('Emisiones de CO2 por año')

# Mostrar la gráfica
plt.show()
```
